````python
import PyPDF2

pdf_file_path = '/content/672f5820e6d0b380d1318509.pdf'

def extract_text_from_pdf(pdf_path):
    text = ""
    try:
        with open(pdf_path, 'rb') as file:
            reader = PyPDF2.PdfReader(file)
            for page_num in range(len(reader.pages)):
                page = reader.pages[page_num]
                text += page.extract_text() if page.extract_text() else ''
    except Exception as e:
        text = f"Error reading PDF: {e}"
    return text

pdf_text = extract_text_from_pdf(pdf_file_path)

print("Extracted text preview (first 2000 characters):")
print(pdf_text[:2000])
````

**Question 1:** Explain the fundamental differences between DDL, DML, and DQL commands in SQL. Provide one example for each type of command.

**Answer:**

*   **DDL (Data Definition Language):** Deals with the schema and descriptions of the database. It is used to create, modify, and delete database objects like tables, views, and indexes.
    *   **Example:** `CREATE TABLE Employees (ID INT PRIMARY KEY, Name VARCHAR(50));`

*   **DML (Data Manipulation Language):** Deals with the manipulation of data stored in the database. It is used to insert, update, and delete records.
    *   **Example:** `INSERT INTO Employees (ID, Name) VALUES (1, 'Alice');`

*   **DQL (Data Query Language):** Primarily used for retrieving data from the database. It consists of only one command, `SELECT`.
    *   **Example:** `SELECT Name FROM Employees WHERE ID = 1;`

**Question 2:** What is the purpose of SQL constraints? Name and describe three common types of constraints, providing a simple scenario where each would be useful.

**Answer:**

**Purpose of SQL Constraints:** SQL constraints are rules enforced on data columns in a table. They are used to limit the type of data that can go into a table, ensuring the accuracy and reliability of the data in the database. This helps maintain data integrity.

**Three Common Types of Constraints:**

1.  **PRIMARY KEY:** Uniquely identifies each row in a table. It must contain unique values, and it cannot contain NULL values. A table can have only one primary key, which can consist of single or multiple columns.
    *   **Scenario:** In an `Orders` table, `OrderID` would be a primary key to uniquely identify each order.

2.  **FOREIGN KEY:** A key used to link two tables together. A foreign key is a field (or collection of fields) in one table that refers to the primary key in another table. It establishes and enforces a link between the data in two tables, ensuring referential integrity.
    *   **Scenario:** In an `Orders` table, `CustomerID` would be a foreign key referencing the `CustomerID` (primary key) in the `Customers` table, ensuring that every order is linked to an existing customer.

3.  **NOT NULL:** Ensures that a column cannot have a NULL value. This means that a column must always contain a value.
    *   **Scenario:** In an `Employees` table, the `LastName` column should be `NOT NULL` because every employee must have a last name.

**Question 3:** Explain the difference between LIMIT and OFFSET clauses in SQL. How would you use them together to retrieve the third page of results, assuming each page has 10 records?

**Answer:**

*   **`LIMIT` Clause:** The `LIMIT` clause is used to restrict the number of rows returned by a query. It specifies the maximum number of rows that the result set should contain.

*   **`OFFSET` Clause:** The `OFFSET` clause is used to skip a specified number of rows from the beginning of the result set before applying the `LIMIT`. It determines the starting point for retrieving rows.

**Retrieving the third page of results (each page has 10 records):**

To retrieve the third page of results, where each page has 10 records, you would use `LIMIT` to specify that you want 10 records, and `OFFSET` to skip the first 20 records (2 pages * 10 records/page).

```sql
SELECT * FROM YourTable
ORDER BY SomeColumn -- ORDER BY is crucial for consistent pagination
LIMIT 10 OFFSET 20;
```

This query would fetch records 21 through 30 (the third page).

**Question 4:** What is a Common Table Expression (CTE) in SQL, and what are its main benefits? Provide a simple SQL example demonstrating its usage.

**Answer:**

**Common Table Expression (CTE):** A CTE is a temporary, named result set that you can reference within a single `SELECT`, `INSERT`, `UPDATE`, or `DELETE` statement. It's essentially a temporary view that exists only for the duration of a single query.

**Main Benefits:**

1.  **Readability and Maintainability:** CTEs break down complex queries into simpler, logical steps, making the SQL code easier to read, understand, and maintain.
2.  **Recursion:** CTEs can be self-referencing (recursive), allowing for complex hierarchical queries (e.g., organizational charts, bill of materials).
3.  **Modularity:** You can define multiple CTEs in a single query, and one CTE can reference another, promoting modularity.
4.  **Avoidance of Subqueries:** They can often replace complex subqueries, leading to more structured and readable code.

**Simple SQL Example:**

Let's say we have an `Orders` table with `CustomerID` and `OrderTotal`, and we want to find the average order total for customers who have placed orders greater than $100.

```sql
WITH HighValueOrders AS (
    SELECT CustomerID, OrderTotal
    FROM Orders
    WHERE OrderTotal > 100
)
SELECT CustomerID, AVG(OrderTotal) AS AverageHighOrderTotal
FROM HighValueOrders
GROUP BY CustomerID;
```

In this example, `HighValueOrders` is a CTE that first filters orders with a total greater than $100. The main query then uses this CTE to calculate the average order total for those specific customers.

**Question 5:** Describe the concept of SQL Normalization and its primary goals. Briefly explain the first three normal forms (1NF, 2NF, 3NF).

**Answer:**

**SQL Normalization:** Normalization is a systematic process of organizing the columns and tables of a relational database to minimize data redundancy and improve data integrity. It involves decomposing a large table into smaller, well-structured tables and defining relationships between them.

**Primary Goals:**

1.  **Eliminate Data Redundancy:** Avoid storing the same piece of data multiple times, which saves storage space and prevents inconsistencies.
2.  **Improve Data Integrity:** Ensure that data is accurate and consistent across the database by preventing update, insertion, and deletion anomalies.
3.  **Simplify Querying:** Make the database structure more logical and easier to query.
4.  **Enhance Database Design:** Create a more robust and flexible database schema.

**First Three Normal Forms:**

1.  **First Normal Form (1NF):**
    *   Eliminates repeating groups in tables (i.e., each cell should contain only a single value).
    *   Each column must contain atomic (indivisible) values.
    *   Each row must be unique.
    *   **Example:** A table of `Orders` should not have a column like `Item1, Item2, Item3`. Instead, items should be in a separate `OrderItems` table linked by `OrderID`.

2.  **Second Normal Form (2NF):**
    *   Must be in 1NF.
    *   All non-key attributes must be fully functionally dependent on the primary key. This means that non-key attributes should not depend on only a part of a composite primary key.
    *   **Example:** If a table has a composite primary key `(OrderID, ItemID)`, and `ItemName` depends only on `ItemID` (not `OrderID`), then `ItemName` should be moved to an `Items` table with `ItemID` as its primary key.

3.  **Third Normal Form (3NF):**
    *   Must be in 2NF.
    *   Eliminates transitive dependencies. This means that non-key attributes should not depend on other non-key attributes.
    *   **Example:** In a `Customers` table, if `ZipCode` determines `City` and `State`, and `City` and `State` are not part of the primary key, then `City` and `State` should be moved to a separate `ZipCodeInfo` table with `ZipCode` as its primary key.

**Question 8:** Retrieve Product Information with Category: Write a SQL query to display the ProductName, Price, StockQuantity, and CategoryName for all products. Order the results by CategoryName and then ProductName alphabetically.

**Answer:**
Assuming there are `Products` and `Categories` tables, and `Products` has a `CategoryID` column that references the `Categories` table's primary key (e.g., `CategoryID`), the query would look like this:

```sql
SELECT
    p.ProductName,
    p.Price,
    p.StockQuantity,
    c.CategoryName
FROM
    Products p
JOIN
    Categories c ON p.CategoryID = c.CategoryID
ORDER BY
    c.CategoryName, p.ProductName ASC;
```

**Question 9:** Write a SQL query that uses a Common Table Expression (CTE) and a Window Function (specifically ROW_NUMBER() or RANK()) to display the CategoryName, ProductName, and Price for the top 2 most expensive products in each CategoryName.

**Answer:**
Assuming the same `Products` and `Categories` table structure as in Question 8:

```sql
WITH RankedProducts AS (
    SELECT
        c.CategoryName,
        p.ProductName,
        p.Price,
        ROW_NUMBER() OVER (PARTITION BY c.CategoryName ORDER BY p.Price DESC) as rn
    FROM
        Products p
    JOIN
        Categories c ON p.CategoryID = c.CategoryID
)
SELECT
    CategoryName,
    ProductName,
    Price
FROM
    RankedProducts
WHERE
    rn <= 2
ORDER BY
    CategoryName, Price DESC;
```

**Question 10 - Task 2:** Which 3 movie categories have the highest rental counts? Display the category name and number of times movies from that category were rented.

**Answer:**
Assuming `Category`, `Film_Category`, `Film`, `Inventory`, and `Rental` tables in the Sakila database:

```sql
SELECT
    c.name AS CategoryName,
    COUNT(r.rental_id) AS RentalCount
FROM
    category c
JOIN
    film_category fc ON c.category_id = fc.category_id
JOIN
    film f ON fc.film_id = f.film_id
JOIN
    inventory i ON f.film_id = i.film_id

    rental r ON i.inventory_id = r.inventory_id
GROUP BY
    c.name
ORDER BY
    RentalCount DESC
LIMIT 3;
```

**Question 10 - Task 3:** Calculate how many films are available at each store and how many of those have never been rented.

**Answer:**
Assuming `Store`, `Inventory`, `Film`, and `Rental` tables in the Sakila database:

```sql
SELECT
    s.store_id,
    COUNT(DISTINCT i.film_id) AS TotalUniqueFilmsAvailable,
    SUM(CASE WHEN r.rental_id IS NULL THEN 1 ELSE 0 END) AS FilmsNeverRented
FROM
    store s
JOIN
    inventory i ON s.store_id = i.store_id
LEFT JOIN
    rental r ON i.inventory_id = r.inventory_id
GROUP BY
    s.store_id
ORDER BY
    s.store_id;
```

**Question 10 - Task 5:** Identify customers who have rented more than 10 times in the last 6 months.

**Answer:**
Assuming `Customer` and `Rental` tables in the Sakila database, and using `rental_date` from the `rental` table:

```sql
SELECT
    c.first_name || ' ' || c.last_name AS CustomerName,
    c.email AS CustomerEmail,
    COUNT(r.rental_id) AS RentalCount
FROM
    customer c
JOIN
    rental r ON c.customer_id = r.customer_id
WHERE
    r.rental_date >= DATE('now', '-6 months') -- For SQLite
    -- r.rental_date >= NOW() - INTERVAL '6 months' -- For PostgreSQL
    -- r.rental_date >= DATE_SUB(NOW(), INTERVAL 6 MONTH) -- For MySQL
GROUP BY
    c.customer_id, CustomerName, CustomerEmail
HAVING
    COUNT(r.rental_id) > 10
ORDER BY
    RentalCount DESC;
```
